0.หารัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุดและน้อยที่สุด

```python
merged_table.groupby('รัฐ').count()  #ตรวจดูจำนวนเที่ยวบินของแต่ละรัฐ
```

```python
max = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินต้นทาง'],ascending=False)  ##หารัฐที่มีขาเข้ามากที่สุด
max.iloc[:1,3:4]
```

```python
min = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินต้นทาง'],ascending=False) ##หารัฐที่มีขาเข้าน้อยที่สุด
min.iloc[-1:,3:4]
```

5.วันธรรมดา หรือวันเสาร์-อาทิตย์มีเครื่องบินขาเข้ามากกว่ากัน (วันจันทร์ = 1, วันอังคาร = 2, ...)

```python
merged_table.groupby('วันในสัปดาห์').count() #ตรวจดูวันในสัปดาห์
```

```python
week = merged_table.groupby('วันในสัปดาห์').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #ดูจำนวนเครื่องบินขาเข้าในแต่ละวัน
week.iloc[:,3:4]
```

```python
Weekday = week.iloc[:5,3:4].sum() #หาจำนวนเครื่องบินขาเข้าในวันธรรมดา
Weekday
```

```python
Weekend = week.iloc[5:7,3:4].sum() #หาจำนวนเครื่องบินขาเข้าในวัน เสาร์-อาทิตย์
Weekend
```

วันธรรมดา มีเครื่องบินขาเข้ามากกว่าวันเสาร์-อาทิตย์
