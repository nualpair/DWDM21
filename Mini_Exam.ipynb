0.หารัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุดและน้อยที่สุด

```python
merged_table['ความล่าช้าขาเข้า'].max()
```

```python
max = merged_table.groupby('รัฐ').max().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False)
max.iloc[:1,:]
```

```python
merged_table['ความล่าช้าขาเข้า'].min()
```

```python
min = merged_table.groupby('รัฐ').min().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False)
min.iloc[-1:,:]
```
