#**เอาเเค่เพศ 1 หรือ เพศ ชาย**
###**เพื่อตัดข้อมูลที่ซ้ำกันของ dcode เป็นการ cleansing ของมูลเพื่อให้ดูง่ายขึ้น**

```python
sex1 = data[data['sex'] == 1 ]
sex1
```

```python
sex2 = data1[data1['sex'] == 1 ]
sex2
```

```python
sex3 =data2[data2['sex'] == 1]
sex3
```

##**เลือกคอลัมที่ต้องการ**
**ต้องการเพศชายที่มี อายุ 20**

```python
sex11 = sex1[['id','dcode','sex','age20',]]
sex11 #ดึงข้อมูลจากปี 56
```

```python
sex22 = sex2[['id','dcode','sex','age20']]
sex22 #ดึงข้อมูลจากปี 59
```

```python
sex33 = sex3[['id','dcode','sex','age20']]
sex33 #ดึงข้อมูลจากตารางของปี 57
```

##**เชื่อมตาราง**

```python
merged_table = sex11.merge(sex22,how='left',left_on='dcode',right_on='dcode')
merged_table ##เชื่อมตารางเเรก ระหว่าง ตาราง sex11 ปี 56 เเละ sex22 ปี 59  ปล.ลองเชื่อด้วย sex เเล้วเเต่เกิด missing ในข้อมูลทั้งที่ในตารางก็มีข้อมูลอยู่เเล้ว ดังนั้นจึงมี dcode ที่เป็น index เพียงตัวด้วย
```

```python
merged_table1 = merged_table.merge(sex33,how='left',left_on='dcode',right_on='dcode')
merged_table1 #เอาตารางล่าสุดที่เชื่อม 2 ตารางเเรก มาเชื่อมกับตาราง sex33 ของปี 57 มาเชื่อม เเละเชื่อมด้วย dcode
```

##**ชื่อคอลัมซ้ำกันเพราะเป็นตารางที่เหมือนกันเราจึงจะเปลี่ยนชื่อตารางเพื่อให้ดูง่ายขึ้น**

```python
merged_table1.rename(columns={'id_x':'id56','sex_x':'sex56','age20_x':'age20_56','id_y':'id59','sex_y':'sex59','age20_y':'age20_59','id':'id57','sex':'sex57','age20':'age20_57'},inplace = True)
merged_table1
```
