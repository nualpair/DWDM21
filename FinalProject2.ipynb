##**รวมตาราง**

```python
merged_table_dcode_sex = data0_X.merge(data1_X,how='left',left_on='dcode_sex',right_on='dcode_sex')
merged_table_dcode_sex  #ครั้งที่ 1 โดยจะให้ตาราง data0_X เป็นตารางเเรก เเละ data1_X เป็นตารางรอง โดยมี index ในการเชื่อม คือ dcode_sex
```
